```python
#slicing
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Train.csv')
slice1=df.loc[20:27,['Warehouse_block','Cost_of_the_Product']]
plt.figure(figsize=(5,5))
plt.bar(slice1['Warehouse_block'],slice1['Cost_of_the_Product'],color = 'green',width = 0.5)
plt.xlabel('WareHouseBlock')
plt.ylabel('Cost of the product')
plt.title('Warehouse and cost of the product ratio')
#plt.tight_layout()
plt.show()
```

```python
#indexing
#same like slicing we can access the single element
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Train.csv')
ind=df[df['Weight_in_gms']>2000]
plt.figure(figsize=(5,5))
plt.bar(ind['Product_importance'],ind['Weight_in_gms'], color = 'green',width = 0.8 )
plt.xlabel('Product_Importance')
plt.ylabel('Weight in gms')
plt.title('Ratio Of Product Importance & Weight in gms')
plt.tight_layout()
plt.show()
```

```python
#Manipulation-Sorting(acending)
#We Can Modify data as per requirement
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Train.csv')
s2=df.loc[15:30,['Mode_of_Shipment','Weight_in_gms']]
s1=s2.sort_values(by='Weight_in_gms',ascending=True)
plt.figure(figsize=(5,5))
#print(s1)
plt.bar(s1['Mode_of_Shipment'],s1['Weight_in_gms'],color='green',width=0.8,align='center')
plt.xlabel('Mode_of_Shipment')
plt.ylabel('Weight_in_gms')
plt.title('Ratio of Mode Of Shipment WRT Weight In Grams ')
plt.tight_layout()
plt.show()
```

```python
#is_Null
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Train2.csv')
s1=df.isnull().sum()
plt.figure(figsize=(25,10))
plt.bar(s1.index,s1.values,color='green',width=0.4)
plt.xlabel('Columns')
plt.ylabel('Missing Values Count')
plt.title('Missing Values Count per Column')
plt.show()
```
